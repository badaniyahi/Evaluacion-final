```python
#@title Pregunta 1 - Impacto de la tecnología de columnas de pared divisoria
messages = [
    {"role"   : "system",
     "content": context1},
    {"role"   : "user",
     "content": """¿Cómo impacta la tecnología de columnas de pared divisoria (DWC) en la eficiencia energética y la reducción de costos operativos en unidades de isomerización de nafta ligera?"""},
]

#What points stand out about data privacy?

prompt = pipeline.tokenizer.apply_chat_template(
        messages,
        tokenize=False,
        add_generation_prompt=True
)

terminators = [
    pipeline.tokenizer.eos_token_id,
    pipeline.tokenizer.convert_tokens_to_ids("<|eot_id|>")
]

outputs = pipeline(
    prompt,
    max_new_tokens=256,
    eos_token_id=terminators,
    do_sample=True,
    temperature=0.6,
    top_p=0.9,
)

print(outputs[0]["generated_text"][len(prompt):])
```
